#### PANDAS 
Used for data analysis, data cleaning. Provides two datastructures series and dataframe.

```python
import pandas as pd
import numpy as np
```

```python
# Pandas Series - One dimensional array-like object that can hold any datatype
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print(f'Series\n{series}')
print(type(series))
```

```python
# Series from a dictionary
data = {'a':1, 'b': 2, 'c':3} # Keys become indexes and values become values
dict_series = pd.Series(data)
print(dict_series)
```

```python
data = [10, 50, 100]
index = ['x', 'y', 'z']
print(pd.Series(data, index))
```

```python
# Dataframe
data = {
  'Name': ['Hardik', 'Virat', 'Rohit'],
  'Age': [30, 37, 38],
  'State': ['Gujarat', 'Delhi', 'Mumbai']
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
print(np.array(data))
```

```python
# Creating a df from a list of dictionaries
data = [
  {'Name': 'Bumrah', 'Age': 30, 'Team': 'Mumbai'},
  {'Name': 'Jitesh', 'Age': 32, 'Team': 'Bangalore'},
  {'Name': 'Arshdeep', 'Age':26 ,'Team': 'Punjab'},
]
df = pd.DataFrame(data)
print(df)
```
