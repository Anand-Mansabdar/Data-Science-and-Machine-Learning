```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Basic plotting with seaborn
df = sns.load_dataset('tips')
df.head()
```

```python
# Scatter plot
sns.scatterplot(x=df['total_bill'], y=df['tip'])
plt.title('Total Bill vs Tips')
```

```python
# Lineplot
sns.lineplot(data=df, x='size', y='total_bill')
plt.title('Total Bill vs Size')
```

```python
# Categorical Plots

# 1. Bar plot
sns.barplot(data=df, x='day', y='total_bill')
plt.title('Bar plot(Total bill vs Day)')
```

```python
# Box plot
sns.boxplot(data=df, x='day', y='total_bill')
```

```python
# Violin Plot
sns.violinplot(data=df, x='day', y='total_bill')
```

```python
# HIstogram
sns.histplot(data=df, x='total_bill', bins=10, kde=True, color='orangered')
```

```python
# KDE plot
sns.kdeplot(data=df, x='total_bill', fill=True,color='red')
```

```python
# Pair plot
sns.pairplot(data=df)
```

```python
# Heatmap
corr = df[['total_bill', 'tip', 'size']].corr()
print(corr)
sns.heatmap(corr)
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
