```python
import sqlite3
```

```python
connect = sqlite3.connect('example.db')
connect
```

```python
cursor = connect.cursor()
```

```python
# Creating a table
cursor.execute('''
Create Table If Not Exists employees(
  id Integer Primary Key,
  name Text not null,
  age Integer,
  department Text
)               
''')

# Commit the changes
connect.commit()
```

```python
cursor.execute('''
Select * from employees             
''')
```

```python
# Inserting data in table
cursor.execute('''
Insert Into employees(name, age, department) values ('Anand', 20, 'Student')
''')

cursor.execute('''
Insert Into employees(name, age, department) values ('Virat', 27, 'Athlete')
''')

cursor.execute('''
Insert Into employees(name, age, department) values ('Rohit', 38, 'Athelete')
''')
```

```python
connect.commit()
```

```python
cursor.execute('Select * from employees')
rows = cursor.fetchall()

# print the data
for row in rows:
  print(row)
```

```python
# Update data in the table
cursor.execute('''
UPDATE employees Set age = 21 where Name='Anand'
''')

connect.commit()
```

```python
cursor.execute('Select * from employees')
rows = cursor.fetchall()

for row in rows:
  print(row)
```

```python
# Deleting data from table
cursor.execute('''Delete from employees where Name='Rohit'
''')
connect.commit()
```
