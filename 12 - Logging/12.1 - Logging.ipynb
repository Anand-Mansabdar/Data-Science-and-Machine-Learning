```python
import logging
# Configuring logging
logging.basicConfig(
  level=logging.DEBUG,
  filename='app.log',
  filemode='w',
  format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
  datefmt='%Y-%M-%d %H:%M:%S'
)
```

```python
logging.debug(msg='This is used for debugging')
logging.info(msg='This is used for some information')
logging.warning(msg='This is used to issue warnings')
logging.error(msg='This is for an error')
logging.critical(msg="This is used for providing critical messages")
```
