```python
# Extracting date, month and year from the Date_of_Journey (Feature Engineering)
df['date'] = df['Date_of_Journey'].str.split('/').str[0]
```

```python
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
```

```python
df['year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df
```

```python
df.info()
```

```python
df['date'] = df['date'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)
df.info()

# Converting date, month and year from object to integer
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minutes'] = df['Arrival_Time'].str.split(':').str[1]
df
```

```python
df['Arrival_Minutes'] = df['Arrival_Minutes'].str.split(' ').str[0]
```

```python
df
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minutes'] = df['Arrival_Minutes'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df
```

```python
df['Departure_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_Minutes'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Departure_Hour'] = df['Departure_Hour'].astype(int)
df['Departure_Minutes'] = df['Departure_Minutes'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
df
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({
  'non-stop':0,
  '1 stop': 1,
  '2 stops': 2,
  '3 stops': 3,
  '4 stops': 4,
  np.nan: 1
})
```

```python
df.head()
```

```python
df['Total_Stops'].isna().sum()
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.columns
```

```python
df['Duration_hours'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df['Duration_Minutes'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df
```

```python
df['Duration_Minutes'] = df['Duration_Minutes'].fillna(0)
df
```

```python
df.drop('Duration', axis=1, inplace=True)
```
