```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
# Creating synthetic datapoints
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
model = SVC(kernel='linear')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
rbf_svc = SVC(kernel='rbf')
rbf_svc.fit(X_train, y_train)
```

```python
y_rbf_pred = rbf_svc.predict(X_test)
```

```python
print(accuracy_score(y_test, y_rbf_pred))
print(confusion_matrix(y_test, y_rbf_pred))
print(classification_report(y_test, y_rbf_pred))
```

```python
poly = SVC(kernel='poly')
poly.fit(X_train, y_train)
y_pred_poly = poly.predict(X_test)
print(confusion_matrix(y_test, y_pred_poly))
print(classification_report(y_test, y_pred_poly))
```

```python
sig = SVC(kernel='sigmoid')
sig.fit(X_train, y_train)
y_pred_sig = poly.predict(X_test)
print(confusion_matrix(y_test, y_pred_sig))
print(classification_report(y_test, y_pred_sig))
```

```python
from sklearn.model_selection import GridSearchCV

# Defining parameter range
param_grid = {
  'C': [0.1, 1, 10, 100, 1000],
  'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
  'kernel': ['rbf']
}
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid, cv=5, verbose=3, refit=True)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
# Prediction
y_pred_grid = grid.predict(X_test)
print(classification_report(y_test, y_pred_grid))
print(confusion_matrix(y_test, y_pred_grid))
```
