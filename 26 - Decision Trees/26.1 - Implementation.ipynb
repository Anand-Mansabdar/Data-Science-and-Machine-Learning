```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris
```

```python
print(iris["DESCR"])
```

```python
print(iris['target'])
```

```python
iris['data']
```

```python
df = pd.DataFrame(iris["data"], columns=["sepal length in cm", "sepal width", "petal length", "petal width"])
```

```python
df
```

```python
# Dependent features
X = df
```

```python
X
```

```python
# Independent Features
y = iris['target']
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn import tree
plt.figure(figsize=(25, 15))
tree.plot_tree(model, filled=True)
```

```python
plt.figure(figsize=(25, 15))
tree.plot_tree(model, filled=True, max_depth=2)
```

```python
param_grid = {
  'criterion': ["gini", "entropy", "log_loss"],
  'splitter': ['best', 'random'],
  'max_depth': [1, 2, 3, 4, 5],
  'max_features': ["int", "float", "auto", "sqrt", "log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(model, param_grid=param_grid, cv=5, scoring="accuracy")
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_grid = grid.predict(X_test)
print(classification_report(y_test, y_pred_grid))
print(accuracy_score(y_test, y_pred_grid))
print(confusion_matrix(y_test, y_pred_grid))
```
