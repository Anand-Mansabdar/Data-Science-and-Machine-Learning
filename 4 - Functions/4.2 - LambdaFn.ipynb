#### LAMBDA FUNCTIONS
Anonymous Functions(Functions without a name), can have any number of arguments but only one expression

```python
def add(a, b):
  return a + b

add(10, 10)
```

```python
addition = lambda a,b: a+b
print(type(addition))
print(addition(30, 50))
```

```python
even = lambda num : num%2==0
print(even(10))
print(even(55))
```

```python
# lambda function with .map() method
nums = [1,2, 3, 4, 5]
squares = list(map(lambda x: x**2, nums))
squares
```
