```python
nums = [1, 2, 3, 4, 5, 6]
def square(x):
  return x**2
list(map(square, nums))
```

```python
# Lambda function with map
list(map(lambda x: x**x, nums))
```

```python
# Map multiple iterables
nums1 = [1, 2, 3]
nums2 = [10, 20, 30]

nums3 = list(map(lambda x, y: x+y, nums1, nums2))
print(nums3)
```

```python
# map() to convert a list if strings to integers
str_nums = ['1', '2', '3', '4', '5']
int_nums = list(map(int, str_nums))

print(int_nums)
```

```python
words = ['apple', 'banana', 'cherry']
upper_words = list(map(str.upper, words))
print(upper_words)
```

```python
def get_name(person):
  return person['name']

people = [{'name': 'Anand', 'age': 20},
          {'name': 'Virat', 'age': 37}]

list(map(get_name, people))
```
