```python
def even(num):
  return num%2==0

print(even(11))

```

```python
l= [1, 2, 3, 4, 5, 6, 7, 8]
list(filter(even, l))
```

```python
nums = [10, 20, 1, 23, 60, 88, 97, 34]
res = list(filter(lambda x: x>50, nums))
res
```

```python
# Fetching nums greater than 50 which are even
res2 = list(filter(lambda x: x>50 and x%2==0, nums))
res2
```

```python
# filter() to check if the age > 25 in a dictionary
people = [{'name': 'Anand', 'age': 20},
          {'name': 'Virat', 'age': 37}]

def age(info):
  return info['age'] > 25

list(filter(age, people))
```
