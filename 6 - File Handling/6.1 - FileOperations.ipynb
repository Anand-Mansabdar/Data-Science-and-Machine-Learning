```python
# Reading a whole file
with open('example.txt', 'r') as file:
  content = file.read()
  print(content)
    
```

```python
# Reading a file line by line
with open('example.txt', 'r') as file:
  for line in file:
    print(line)
```

```python
# Reading a file line by line
with open('example.txt', 'r') as file:
  for line in file:
    print(line.strip())
```

```python
# Writing into a file
with open('example.txt','w') as file:
  file.write('I like playing cricket!\n')
  file.write('Currently learning data science')
  
# The above code completely overwrites a file ie, the content previously present will be removed
```

```python
# Writing into a file without overwriting it
with open('example.txt', 'a') as file:
  file.write('\nHello. My name is Anand')
  file.write('\nMy favourite crickter is Virat Kohli')
# 'a' stands for append
```

```python
# Writing a list of lines to a file
lines = ['First line\n', 'Second line\n', 'Third line\n']
with open('newfile.txt', 'a') as file:
  file.writelines(lines)
```

```python
# Binary Files

# Writing into a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
  file.write(data)
```

```python
# Reading a binary file
with open('example.bin', 'rb') as file:
  content = file.read()
  print(content)
```

```python
# Read the content from a source file and write it into a destination file

with open('example.txt', 'r') as source_file:
  content = source_file.read()
  
with open('destination.txt', 'w') as dest_file:
  dest_file.write(content)
```

```python
# Read a file and count the number of lines, words, and characters.
# Counting lines, words, and characters in a text file
def count_text_file(file_path):
  with open(file_path, 'r') as file:
    lines = file.readlines()
    line_count = len(lines)
    word_count = sum(len(line.split()) for line in lines)
    char_count = sum(len(line) for line in lines)
  return line_count, word_count, char_count

file_path = 'example.txt'
lines, words, charcters =count_text_file(file_path)
print(f'Lines: {lines}, Words: {words}, Charcters: {charcters}')
```

```python
# Writing and then reading : w+ mode is used for both reading and writing. If the file does not exist, it will be created. If the file exists then its content will be overwritten
with open('example.txt', 'w+') as file:
  file.write('Hello World')
  file.write('This is a new line')
  
  # Mode the file cursor to the beginning
  file.seek(0)
  
  # Read the content of the file
  content = file.read()
  print(content)
```
