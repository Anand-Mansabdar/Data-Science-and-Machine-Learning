```python
# creating a new directory
import os
new_dir = 'package'
os.mkdir(new_dir)
print(f'Directory {new_dir} created')
```

```python
# Listing files and directories
items = os.listdir('.') # '.' means we are looking in the current working directory
print(items)
```

```python
# Joining paths
dir_name = 'folder'
file_name = 'file.txt'
full_path = os.path.join(dir_name, file_name)
print(full_path)
```

```python
# With the name of current working directory
dir_name = 'folder'
file_name = 'file.txt'
full_path = os.path.join(os.getcwd(),dir_name, file_name)
print(full_path)
```

```python
path = 'example.txt'
if os.path.exists(path):
  print(f'The path {path} exists')
else: 
  print(f'The path {path} does not exist')
  
```

```python
# Checking if a path is a file or a directory
path = 'package'
if os.path.isfile(path):
  print(f'{path} a file')
elif os.path.isdir(path):
  print(f'{path} a directory')
else:
  print(f'{path} is neither a file or a directory')
```

```python
# Getting the absolute path
relative_path = 'example.txt'
absolute_path = os.path.abspath(relative_path)
print(absolute_path)
```
