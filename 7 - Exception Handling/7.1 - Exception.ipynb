```python
try:
  num = int(input('Enter a number: '))
  result = 10/num
except ValueError:
  print(f'This is not a valid number')
except ZeroDivisionError:
  print(f'Cannot divide by zero')
except Exception as ex:
  print(ex)
```

```python
# Try except and else block
try:
  num = int(input('Enter a number: '))
  result = 10/num
except ValueError:
  print(f'This is not a valid number')
except ZeroDivisionError:
  print(f'Cannot divide by zero')
except Exception as ex:
  print(ex)
else: # This block is executed only if there are no errors are occured in the code
  print(result)
  print('Program executed without an error')
```

```python
# Try, except, else and finally
try:
  num = int(input('Enter a number: '))
  result = 10/num
except ValueError:
  print(f'This is not a valid number')
except ZeroDivisionError:
  print(f'Cannot divide by zero')
except Exception as ex:
  print(ex)
else:
  print(result)
  print('Program executed without an error')
finally:
  print('Finally block is executed') # This block gets executed in case the error has occured ot not
```

```python
# Practical Examples

# File handling and exception handling
try:
  file = open('example.txt', 'r')
  content = file.read()
  print(content)
except FileNotFoundError as ex:
  print(ex)
  print('The file does not exist')
finally:
  if 'file' in locals() or not file.closed():
    file.close()
    print('File closed')
```
