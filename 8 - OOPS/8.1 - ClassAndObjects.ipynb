```python
# Instance Variable and Methods
class Dog:
  # Constructor :- we use it to define all the attributes that we want to put in out class
  def __init__(self, name, age):
    self.name = name # self.name is the instance variable of our class
    self.age = age

dog1 = Dog('jazz', 4)
print(dog1)
print(f'Dog1 --> Name: {dog1.name}, Age: {dog1.age}')

dog2 = Dog('Charlie', 6)
print(f'Dog2 --> Name: {dog2.name}, Age: {dog2.age}')
```

```python
# Instance methods
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def greet(self):
    print(f'{self.name} is greeting you !!!')
    
human1 = Human('Virat Kohli', 37)
print(human1.age)
human1.greet()

human2 = Human('Anand Mansabdar', 20)
print(human2.name)
human2.greet()
```

```python
# Modelling a back account
class Bank:
  def __init__(self, account_name, balance=0):
    self.account_name = account_name
    self.balance = balance
  
  def deposit(self, amount):
    if(amount <= 0):
      print('You cannot add a negative amount or a 0 amount to your bank account')
      return
    self.balance += amount
    print(f'{amount} is deposited in your account. Your new balance is {self.balance}')
    
  def withdraw(self, amount):
    if amount > self.balance:
      print('Insufficient funds')
    else:
      self.balance -= amount
      print(f'{amount} has been withdrawn from your account. Remaining Balance is {self.balance}')
    


account_1 = Bank('Anand_Mansabdar')
account_1.deposit(10000)
account_1.deposit(5000)
account_1.deposit(2000)
account_1.balance
account_1.withdraw(4300)
account_1.balance
```
