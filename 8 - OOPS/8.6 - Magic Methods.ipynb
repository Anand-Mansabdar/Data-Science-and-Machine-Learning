They are called as dunder methods(double underscore). They are predefined methods in python that you can override to change the behaviour of your objects

```python
'''
--> __init__: Initializes a new instance of a class
--> __str__: Returns a string representation of an object
--> __repr__: Returns an official string represntation of an object
--> __len__: Returns a length of an object
--> __getitem__: Gets an item from a container 
--> __setItem__: Sets an item in a comtainer
'''
```

```python
# Basic methods
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age
    
  def __str__(self):
    return f'{self.name} is {self.age} years old'
  
  def __repr__(self):
    return f'Person(name={self.name}, age={self.age})'
  
  
person = Person('Anand', 20)
print(person)
print(repr(person))
```

```python
# Operator Overloading
class Vector:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  
  def __add__(self, other):
    return Vector(self.x + other.x, self.y + other.y)
   
  def __sub__(self, other):
    return Vector(self.x - other.x, self.y - other.y)
  
  def __mul__(self, other):
    return Vector(self.x * other.x, self.y * other.y)
  
  def __eq__(self, other):
    return self.x == other.x and self.y == other.y
  
  def __repr__(self):
    return f'Vector({self.x}, {self.y})'
  
v1 = Vector(10, 20)
v2 = Vector(30, 40)
print(v1 + v2)
print(v1 - v2)
print(v1 * v2)
print(v1 == v2)

  
```
