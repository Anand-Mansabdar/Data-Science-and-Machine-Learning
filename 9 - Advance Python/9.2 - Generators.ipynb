```python
def my_generator():
  yield 1
  yield 2
  yield 3
  
gen = my_generator()
print(gen)
next(gen)
```

```python
for val in gen:
  print(val)
```

```python
# Practical :- Reading large files
def read_file(file_name):
  with open(file_name, 'r') as file:
    for line in file:
      yield line
 
```

```python
file_name = 'largeFile.txt'
for line in read_file(file_name):
  print(line.strip())
```
