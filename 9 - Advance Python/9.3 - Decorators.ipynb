```python
# Function copy
def welcome():
  return 'Welcome! Good Evening...'

welcome()
```

```python
wel = welcome()
wel
del welcome
```

```python
wel  # When we create a function copy and delete the original one we still have access to the copy function. 
# Even after deleting welcome() we can still use wel and get the result
```

```python
# CLosures
def main_welcome():
  msg = 'Welcome'
  def sub_welcome():
    print(f'Good evening...')
    print(msg)
    print(f'Please come in...')
  return sub_welcome()

main_welcome()
```

```python
def main_welcome(func):
  def sub_welcome():
    print(f'Good evening...')
    func()
    print(f'Please come in...')
  return sub_welcome()

main_welcome(print)
```

```python
def main_welcome(func):
  def sub_welcome():
    print(f'Good evening...')
    func('I am currently learning Data Science')
    print(f'Please come in...')
  return sub_welcome()

main_welcome(print)
```

```python
def main_welcome(func, lst):
  def sub_welcome():
    print(f'Good evening...')
    print(func(lst))
    print(f'Please come in...')
  return sub_welcome()

main_welcome(len,[1, 2, 3, 4, 5])
```

```python
# Decorators
def main_welcome(func):
  def sub_welcome():
    print(f'Good evening...')
    func()
    print(f'Please come in...')
  return sub_welcome()

main_welcome(print)
```

```python
def course_introduction():
  print('I am learning data science')
  
course_introduction()
```

```python
main_welcome(course_introduction)
```

```python
@main_welcome
def course_introduction():
  print('I am learning data science')
```

```python
def greet(fn):
  print('Good evening everyone!!!')
  fn()
  
def intro():
  print('My name is Anand Mansabdar. I am 20 years old and currently pursuing my B.Tech in Computer Science Engineering at Swami Vivekananda Institute of Technology in Hyderabad')
  
```

```python
@greet
def intro():
  print('My name is Anand Mansabdar. I am 20 years old and currently pursuing my B.Tech in Computer Science Engineering at Swami Vivekananda Institute of Technology in Hyderabad')
```

```python
def my_decorator(func):
  def wrapper():
    print("This is being executed before func")
    func()
    print('This is being executed after the func')
  return wrapper
```

```python
@my_decorator
def say_hello():
  print('Hello')
```

```python
say_hello()
```

```python
# Decorators with arguments
def repeat(n):
  def decorator(fn):
    def wrapper(*args, **kwargs):
      for _ in range(n):
        fn(*args, ** kwargs)
    return wrapper
  return decorator

@repeat(10)
def say_hello():
  print('Hello')
```

```python
say_hello()
```
